```python
%matplotlib inline
```

```python
# import dependencies 
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot 
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston', color="blue", linewidth = 2, marker = "*")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot and add a label for the legend.
fig,ax = plt.subplots(figsize=(6,4))
ax.plot(x_axis, y_axis, label='Boston', color="green", linewidth = 2, marker = "D")
# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)
# Create a title.
ax.set_title("PyBer Fare by Month")
# Add a grid.
ax.grid()
# Add the legend.
ax.legend()
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# create a horizontal bar chart 
plt.barh(x_axis, y_axis, label = "Boston", color = "magenta")
plt.gca().invert_yaxis()
plt.legend()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color = "black", label = "New York")
ax.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = "cyan", label = "Chicago")
plt.gca().invert_yaxis()
plt.xlabel("Fare($)")
plt.ylabel("Date")
ax.legend()
```

```python
# Create scatter plot using plt.plot
plt.plot(x_axis, y_axis, 'o')
```

```python
# Create scatter plot using plt.plot
plt.scatter(y_axis, x_axis, color = "red", label = "Chicago")
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.legend()
```

```python
# create a scatter plot chart 
plt.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis])
```

```python
# create a scatter plot chart in object-oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# create a bubble chart in object-oriented interface
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], color = "skyblue", alpha = 0.8, edgecolors = "black", linewidths = 2, label = "Boston")
ax.legend()
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.xlim(0,50)
plt.xlabel("Fare($)")
plt.ylabel("Date")
```

```python
# create a pie chart 
plt.subplots(figsize=(8,8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, 
        labels=x_axis, 
        explode = explode_values, 
        colors = colors,
        autopct='%1.1f%%')
plt.show()
```

```python
# create a pie chart using object-oriented interface
explode = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis, 
      labels = x_axis,
      autopct = '%1.1f%%',
      explode = explode, 
      colors = colors,
      startangle = 270,
      shadow = True 
      )
plt.gca().invert_yaxis()
plt.show()
```
